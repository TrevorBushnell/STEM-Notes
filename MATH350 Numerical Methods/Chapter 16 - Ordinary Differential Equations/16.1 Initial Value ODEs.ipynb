# 16.1: Initial Value ODEs

A **differential equation** is an equation involving an unknown function and its derivatives. It could be only one derivative or multiple derivatives. Here are some examples of differential equations:

* $\frac{dy}{dt} = y + 2t$
    * $y(t)$ is the uknown function and $t$ is the independent variable
* $y' = y^2$
    * *autonomous equation* - the independent variable doesn't explicity show up
    * 1st order ODE because the highest derivative is the first derivative
* $y'' + 3y' + y = \cos(2t)$
    * *linear homogenous equation*
    * 2nd order ODE because the highest derivative is the second derivative
    
An **ordinary differential equation** is a differential equation involving functions with only one variable. A differential equation involving functions with more than one independent variable is called a **partial differential equations**. Some examples of PDEs are as follows:

* $u = u(x,y)$ and $\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0$

In this class, we will just stick to ODEs.

## Solving ODEs

Solving a differential equation means finding the unknown function(s) that satisfy the equation.

> **Example:** Verify that $y(t) = e^t - t$ is a solution to the differential equation $y' = y + t - 1$. 

All we have to do is plug in $y$ into both sides of the equation. Note that $y' = e^t - 1$:

$$
e^t - 1 = e^t - t + t - 1 = e^t - 1
$$

Therefore the function $y = e^t - t$ is a solution to this equation.

There are a few equations that are really straightfoward to solve by hand:


### Separable ODEs

If you have a separable ODE, then you can write the ODE in the form $\frac{dy}{dt} = f(y)\:g(t) \implies \int \frac{1}{f(y)}\:dy = \int g(t)\:dt$. Then you can solve this equation by just solving both integrals.

> **Example:** Solve $\frac{dy}{dt} = 2yt$

$$
\begin{aligned}
    \frac{dy}{dt} &= 2yt\\
    \int \frac{1}{y} \: dy &= \int 2t \: dt\\
    \ln|y| &= t^2 + C\\
    |y| &= e^{t^2 + C}\\
    y &= \pm e^Ce^{t^2}\\
    y &= Ke^{t^2}
\end{aligned}
$$

The solution that we just got is called a "general solution". 

We can turn this problem into an "initial value problem" by adding some sort of initial condition so that you can solve for the $K$ constant. For example, if we knew that $y(0) = 2$, then we could plug these values into our general solution and solve for $K$ so that we get some specific solution.



### Newton's Law of Cooling

Netwon's Law of Cooling is as follows:

$$
\frac{dy}{dt} = k(y-A)\:\:\:\:y(0) = y_0
$$

where $y(t)$ is the temperature as a function of $t$, $y(0) = y_0$ is the initial temperature, and $k$ is the cooling rate such that $k < 0$. If we solve this differential equation, then we get the following:

$$
\begin{aligned}
    y(t) = (y_0 - A)e^{kt} + A
\end{aligned}
$$

## Messing Around with ODEs and Newton's Law of Cooling in MATLAB

Here's some code to play around with Newton's Law of Cooling:

```matlab
k = -0.1;
A = 70;
y0 = 33:10:211;
tspan = [0,2*180];

yprime = @(t,y) k * (y - A);

fig = figure
hold on

for j = 1:numel(y0)
    [t,y] = ode45(yprime, tspan, y0(j));
    plot(t,y)
end

ylabel('Tempreature')
xlabel('Time in Minutes')
title('Solutions to Newtons Law of Cooling with Various Initial Conditions')

hold off
```

If we slow down the cooling rate, then this happens:

```matlab
k = -0.01;
A = 70;
y0 = 33:10:211;
tspan = [0,2*180];

yprime = @(t,y) k * (y - A);

fig = figure
hold on

for j = 1:numel(y0)
    [t,y] = ode45(yprime, tspan, y0(j));
    plot(t,y)
end

ylabel('Tempreature')
xlabel('Time in Minutes')
title('Solutions to Newtons Law of Cooling with Various Initial Conditions')

hold off
```
