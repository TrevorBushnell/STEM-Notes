## Graphing Solutions to PDEs

This MATLAB code will plot a solution to a PDE as a surface, as well as some additional time snapshots. We will solve the following IVP:

$$
u_t +m 2u_{xx} - 0, \:\:\: u(x,0) = e^{-x^2}
$$

```matlab
u = @(x,t) exp(-(x-2*t).^2);

% Set up space-time domain grid
x_max = 4;
t_max = 1;

x = -x_max:(x_max/40):x_max; % Spacial domain
t = 0:(t_max/20):t_max; % temporal domain
[X,T] = meshgrid(x,t); % Space-time grid

% Define u on the space-time grid
U = u(X,T);

% Plot the solution as a surface over the space-time domain
surf(X,T,U)
xlabel('x')
ylabel('t')
zlabel('u')
title('Solution u(x,t) plotted as a surface')
```

```matlab
% Plot some individual time snapshots 
t0 = 0;
plot(x, u(x,t0)); 
xlabel('x'); ylabel('u');
title('Time snapshot of u(x,t) at t=0')
```

```matlab
% Plot some more individual time snapshots

t0 = 1;
plot(x, u(x,t0)); 
xlabel('x'); ylabel('u');
title('Time snapshot of u(x,t) at t=1')
```

```matlab
% Plot multiple time snapshots on the same axes
t0 = 0:0.25:1;
plot(x, u(x,t0(1)), x, u(x,t0(2)), x, u(x,t0(3)), x, u(x,t0(4)), x, u(x,t0(5))); 
xlabel('x'); ylabel('u');
legend('t=0', 't=0.25', 't=0.5', 't=0.75', 't=1'); 
title('Time snapshots of u(x,t)')
```
